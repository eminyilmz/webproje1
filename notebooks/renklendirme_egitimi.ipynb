## Veri Seti

MS-COCO 2017 Train seti kullanılmıştır. 118.287 adet farklı sahnelere ait
gerçek dünya fotoğrafı içerir. Renklendirme görevi için geniş çeşitliliğe
sahip olması bu veri setini ideal kılar.

```python
import urllib.request, zipfile

DATA_DIR = '/content/train2017'
if not os.path.exists(DATA_DIR):
    print('COCO 2017 Train seti indiriliyor (~18GB)...')
    urllib.request.urlretrieve(
        'http://images.cocodataset.org/zips/train2017.zip',
        '/content/train2017.zip'
    )
    print('Arşiv açılıyor...')
    with zipfile.ZipFile('/content/train2017.zip', 'r') as z:
        z.extractall('/content')
    os.remove('/content/train2017.zip')

all_imgs = sorted(glob.glob(os.path.join(DATA_DIR, '*.jpg')))
print(f'Toplam fotoğraf: {len(all_imgs):,}')
```

```python
# Eğitim / doğrulama ayrımı (%95 / %5)
split = int(len(all_imgs) * 0.95)
train_paths = all_imgs[:split]
val_paths   = all_imgs[split:]

IMG_SIZE = 256

class ColorizationDataset(Dataset):
    """
    RGB fotoğrafları L*a*b* renk uzayına dönüştürür.
    Girdi: L kanalı (parlaklık — gri tonlama)
    Hedef: ab kanalları (renk bilgisi)
    """
    def __init__(self, paths, size=256):
        self.paths = paths
        self.size = size

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        try:
            img = Image.open(self.paths[idx]).convert('RGB')
            img = img.resize((self.size, self.size), Image.BICUBIC)
            img_np = np.array(img) / 255.0
            lab = rgb2lab(img_np).astype('float32')
            L  = torch.from_numpy((lab[:,:,0:1] / 50.0 - 1.0).transpose(2,0,1))
            ab = torch.from_numpy((lab[:,:,1:3] / 128.0).transpose(2,0,1))
            return L, ab
        except:
            return self.__getitem__((idx + 1) % len(self.paths))

train_dl = DataLoader(ColorizationDataset(train_paths, IMG_SIZE),
                      batch_size=32, shuffle=True, num_workers=4, pin_memory=True)
val_dl   = DataLoader(ColorizationDataset(val_paths, IMG_SIZE),
                      batch_size=32, shuffle=False, num_workers=2, pin_memory=True)

print(f'Eğitim: {len(train_paths):,} fotoğraf ({len(train_dl)} batch)')
print(f'Doğrulama: {len(val_paths):,} fotoğraf ({len(val_dl)} batch)')
```

## Model Mimarisi

U-Net tabanlı encoder-decoder yapısı kullanılmıştır. Skip connection'lar
sayesinde uzamsal bilgi kaybı önlenir. Encoder'da LeakyReLU, decoder'da
ReLU aktivasyon fonksiyonu tercih edilmiştir.

```python
class UNetBlock(nn.Module):
    def __init__(self, in_ch, out_ch, down=True, dropout=False):
        super().__init__()
        layers = []
        if down:
            layers += [nn.Conv2d(in_ch, out_ch, 4, 2, 1, bias=False),
                       nn.BatchNorm2d(out_ch), nn.LeakyReLU(0.2, True)]
        else:
            layers += [nn.ConvTranspose2d(in_ch, out_ch, 4, 2, 1, bias=False),
                       nn.BatchNorm2d(out_ch), nn.ReLU(True)]
            if dropout:
                layers.append(nn.Dropout(0.5))
        self.block = nn.Sequential(*layers)

    def forward(self, x):
        return self.block(x)


class ColorizationUNet(nn.Module):
    def __init__(self):
        super().__init__()
        # Encoder
        self.enc1 = nn.Sequential(nn.Conv2d(1, 64, 4, 2, 1), nn.LeakyReLU(0.2, True))
        self.enc2 = UNetBlock(64, 128)
        self.enc3 = UNetBlock(128, 256)
        self.enc4 = UNetBlock(256, 512)
        self.enc5 = UNetBlock(512, 512)
        self.enc6 = UNetBlock(512, 512)

        # Bottleneck
        self.bottleneck = nn.Sequential(nn.Conv2d(512, 512, 4, 2, 1), nn.ReLU(True))

        # Decoder
        self.dec1 = UNetBlock(512, 512, down=False, dropout=True)
        self.dec2 = UNetBlock(1024, 512, down=False, dropout=True)
        self.dec3 = UNetBlock(1024, 512, down=False)
        self.dec4 = UNetBlock(1024, 256, down=False)
        self.dec5 = UNetBlock(512, 128, down=False)
        self.dec6 = UNetBlock(256, 64, down=False)
        self.final = nn.Sequential(nn.ConvTranspose2d(128, 2, 4, 2, 1), nn.Tanh())

    def forward(self, x):
        e1 = self.enc1(x);  e2 = self.enc2(e1); e3 = self.enc3(e2)
        e4 = self.enc4(e3); e5 = self.enc5(e4); e6 = self.enc6(e5)
        b = self.bottleneck(e6)
        d1 = self.dec1(b)
        d2 = self.dec2(torch.cat([d1, e6], 1))
        d3 = self.dec3(torch.cat([d2, e5], 1))
        d4 = self.dec4(torch.cat([d3, e4], 1))
        d5 = self.dec5(torch.cat([d4, e3], 1))
        d6 = self.dec6(torch.cat([d5, e2], 1))
        return self.final(torch.cat([d6, e1], 1))


model = ColorizationUNet().to(DEVICE)
print(f'Parametre sayısı: {sum(p.numel() for p in model.parameters()):,}')
```

```python
# Perceptual loss (VGG19 feature matching)
class PerceptualLoss(nn.Module):
    def __init__(self):
        super().__init__()
        vgg = models.vgg19(pretrained=True).features[:16].eval()
        for p in vgg.parameters(): p.requires_grad = False
        self.vgg = vgg.to(DEVICE)
        self.l1 = nn.L1Loss()

    def forward(self, pred, target):
        # ab → 3 kanal (L sabit, ab değişken)
        pred_rgb = torch.cat([torch.zeros_like(pred[:,:1]), pred], dim=1)
        tgt_rgb  = torch.cat([torch.zeros_like(target[:,:1]), target], dim=1)
        return self.l1(self.vgg(pred_rgb), self.vgg(tgt_rgb))

perceptual_loss = PerceptualLoss()
```

```python
# Eğitim döngüsü
EPOCHS = 30
LR = 2e-4

l1_criterion = nn.L1Loss()
optimizer = optim.Adam(model.parameters(), lr=LR, betas=(0.5, 0.999))
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

train_losses, val_losses = [], []
best_val = float('inf')
t0 = time.time()

print(f'Eğitim başlıyor — {EPOCHS} epoch, {len(train_dl)} batch/epoch')
print('=' * 70)

for epoch in range(1, EPOCHS + 1):
    model.train()
    running = 0.0
    for i, (L, ab) in enumerate(train_dl):
        L, ab = L.to(DEVICE), ab.to(DEVICE)
        pred = model(L)

        loss = l1_criterion(pred, ab) + 0.1 * perceptual_loss(pred, ab)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        running += loss.item()

        if (i+1) % 500 == 0:
            print(f'  [{epoch}/{EPOCHS}] Batch {i+1}/{len(train_dl)} — loss: {loss.item():.5f}')

    avg_train = running / len(train_dl)
    train_losses.append(avg_train)

    # Doğrulama
    model.eval()
    vl = 0.0
    with torch.no_grad():
        for L, ab in val_dl:
            L, ab = L.to(DEVICE), ab.to(DEVICE)
            vl += l1_criterion(model(L), ab).item()
    avg_val = vl / len(val_dl)
    val_losses.append(avg_val)
    scheduler.step()

    if avg_val < best_val:
        best_val = avg_val
        torch.save(model.state_dict(), '/content/best_colorization.pth')

    elapsed = (time.time() - t0) / 60
    print(f'Epoch [{epoch:2d}/{EPOCHS}]  '
          f'train: {avg_train:.5f}  val: {avg_val:.5f}  '
          f'lr: {optimizer.param_groups[0]["lr"]:.6f}  '
          f'süre: {elapsed:.0f}dk')

print('=' * 70)
print(f'Eğitim tamamlandı. En iyi val loss: {best_val:.5f}')
```

```python
# Eğitim grafiği
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(train_losses, 'b-', label='Eğitim', linewidth=2)
ax.plot(val_losses, 'r--', label='Doğrulama', linewidth=2)
ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
ax.set_title('Renklendirme Modeli — Eğitim Süreci', fontweight='bold')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/egitim_grafigi.png', dpi=150)
plt.show()
```

```python
# Sonuçları görselleştir
model.load_state_dict(torch.load('/content/best_colorization.pth', map_location=DEVICE))
model.eval()

fig, axes = plt.subplots(4, 3, figsize=(14, 18))
test_indices = np.linspace(split, len(all_imgs)-1, 4, dtype=int)

for row, idx in enumerate(test_indices):
    img = Image.open(all_imgs[idx]).convert('RGB').resize((256, 256))
    img_np = np.array(img) / 255.0
    lab = rgb2lab(img_np).astype('float32')
    L = lab[:,:,0:1]
    L_t = torch.from_numpy((L/50-1).transpose(2,0,1)).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        pred_ab = model(L_t).squeeze(0).cpu().numpy().transpose(1,2,0) * 128

    result = np.clip(lab2rgb(np.concatenate([L, pred_ab], 2)), 0, 1)
    gray = np.stack([img_np.mean(2)]*3, 2)

    axes[row,0].imshow(gray);   axes[row,0].set_title('Giriş (S/B)' if row==0 else '')
    axes[row,1].imshow(result); axes[row,1].set_title('Model Çıktısı' if row==0 else '')
    axes[row,2].imshow(img_np); axes[row,2].set_title('Orijinal' if row==0 else '')

for ax in axes.flatten(): ax.axis('off')
plt.suptitle('Renklendirme Sonuçları', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('/content/renklendirme_sonuclari.png', dpi=150, bbox_inches='tight')
plt.show()
```
