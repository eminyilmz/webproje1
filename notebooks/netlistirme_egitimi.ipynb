## Eğitim Konfigürasyonu

Aşağıdaki konfigürasyon, Real-ESRGAN'ın orijinal eğitim pipeline'ını takip eder:
- **İki aşamalı bozulma (degradation)** simülasyonu
- **L1 + VGG Perceptual + GAN** loss kombinasyonu
- **EMA** (Exponential Moving Average) ile model stabilizasyonu

```python
%%writefile /content/Real-ESRGAN/options/finetune_piksel.yml
name: piksel_sr_finetune
model_type: RealESRGANModel
scale: 4
num_gpu: 1
manual_seed: 42

datasets:
  train:
    name: COCO2017
    type: RealESRGANDataset
    dataroot_gt: /content/train2017
    meta_info: /content/Real-ESRGAN/meta_info_coco.txt
    io_backend:
      type: disk

    blur_kernel_size: 21
    kernel_list: ['iso', 'aniso', 'generalized_iso', 'generalized_aniso', 'plateau_iso', 'plateau_aniso']
    kernel_prob: [0.45, 0.25, 0.12, 0.03, 0.12, 0.03]
    sinc_prob: 0.1
    blur_sigma: [0.2, 3]
    betag_range: [0.5, 4]
    betap_range: [1, 2]

    blur_kernel_size2: 21
    kernel_list2: ['iso', 'aniso', 'generalized_iso', 'generalized_aniso', 'plateau_iso', 'plateau_aniso']
    kernel_prob2: [0.45, 0.25, 0.12, 0.03, 0.12, 0.03]
    sinc_prob2: 0.1
    blur_sigma2: [0.2, 1.5]
    betag_range2: [0.5, 4]
    betap_range2: [1, 2]

    final_sinc_prob: 0.8
    gt_size: 256
    use_hflip: true
    use_rot: false

    resize_prob: [0.2, 0.7, 0.1]
    resize_range: [0.15, 1.5]
    gaussian_noise_prob: 0.5
    noise_range: [1, 30]
    poisson_scale_range: [0.05, 3]
    gray_noise_prob: 0.4
    jpeg_range: [30, 95]

    second_blur_prob: 0.8
    resize_prob2: [0.3, 0.4, 0.3]
    resize_range2: [0.3, 1.2]
    gaussian_noise_prob2: 0.5
    noise_range2: [1, 25]
    poisson_scale_range2: [0.05, 2.5]
    gray_noise_prob2: 0.4
    jpeg_range2: [30, 95]

    use_shuffle: true
    num_worker_per_gpu: 2
    batch_size_per_gpu: 4
    dataset_enlarge_ratio: 1
    prefetch_mode: ~

network_g:
  type: RRDBNet
  num_in_ch: 3
  num_out_ch: 3
  num_feat: 64
  num_block: 23
  num_grow_ch: 32

network_d:
  type: UNetDiscriminatorSN
  num_in_ch: 3
  num_feat: 64
  skip_connection: true

path:
  pretrain_network_g: /content/Real-ESRGAN/weights/RealESRGAN_x4plus.pth
  param_key_g: params_ema
  strict_load_g: true
  pretrain_network_d: /content/Real-ESRGAN/weights/RealESRGAN_x4plus_netD.pth
  param_key_d: params
  strict_load_d: true
  resume_state: ~

train:
  ema_decay: 0.999
  optim_g:
    type: Adam
    lr: !!float 1e-4
    weight_decay: 0
    betas: [0.9, 0.99]
  optim_d:
    type: Adam
    lr: !!float 1e-4
    weight_decay: 0
    betas: [0.9, 0.99]

  scheduler:
    type: MultiStepLR
    milestones: [5000, 10000]
    gamma: 0.5

  total_iter: 15000
  warmup_iter: -1

  pixel_opt:
    type: L1Loss
    loss_weight: 1.0
    reduction: mean
  perceptual_opt:
    type: PerceptualLoss
    layer_weights:
      conv1_2: 0.1
      conv2_2: 0.1
      conv3_4: 1
      conv4_4: 1
      conv5_4: 1
    vgg_type: vgg19
    use_input_norm: true
    perceptual_weight: 1.0
    style_weight: 0
    range_norm: false
    criterion: l1
  gan_opt:
    type: GANLoss
    gan_type: vanilla
    real_label_val: 1.0
    fake_label_val: 0.0
    loss_weight: !!float 1e-1

  net_d_iters: 1
  net_d_init_iters: 0

logger:
  print_freq: 200
  save_checkpoint_freq: 5000
  use_tb_logger: true
  wandb:
    project: ~
    resume_id: ~

dist_params:
  backend: nccl
  port: 29500
```

```python
# Sonuçları görselleştir
import matplotlib.pyplot as plt
import cv2, numpy as np
from basicsr.archs.rrdbnet_arch import RRDBNet
from realesrgan import RealESRGANer

ckpts = sorted(glob.glob('/content/Real-ESRGAN/experiments/piksel_sr_finetune/models/net_g_*.pth'))
latest = ckpts[-1] if ckpts else W_G
print(f'Kullanılan checkpoint: {os.path.basename(latest)}')

model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
upsampler = RealESRGANer(scale=4, model_path=latest, model=model, tile=400,
                         tile_pad=10, pre_pad=0, half=True, device='cuda')

test_imgs = sorted(glob.glob('/content/train2017/*.jpg'))[-3:]
fig, axes = plt.subplots(3, 3, figsize=(16, 16))

for row, path in enumerate(test_imgs):
    img = cv2.imread(path)
    h, w = img.shape[:2]
    small = cv2.resize(img, (w//4, h//4), interpolation=cv2.INTER_CUBIC)
    bicubic = cv2.resize(small, (w, h), interpolation=cv2.INTER_CUBIC)
    sr, _ = upsampler.enhance(small, outscale=4)

    axes[row, 0].imshow(cv2.cvtColor(bicubic, cv2.COLOR_BGR2RGB))
    axes[row, 0].set_title('Düşük Çözünürlük (Bicubic)' if row==0 else '')
    axes[row, 1].imshow(cv2.cvtColor(sr[:h,:w], cv2.COLOR_BGR2RGB))
    axes[row, 1].set_title('Model Çıktısı (4x SR)' if row==0 else '')
    axes[row, 2].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axes[row, 2].set_title('Orijinal (Ground Truth)' if row==0 else '')

for ax in axes.flatten(): ax.axis('off')
plt.suptitle('Süper Çözünürlük Eğitim Sonuçları', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/sr_sonuclari.png', dpi=150, bbox_inches='tight')
plt.show()
```
